```python
%%manim -v WARNING --disable_caching -qh VFscene

class VFscene(MovingCameraScene):
    def construct(self):
        func = lambda pos: (pos[0] * np.array([0,1,0]) + pos[1] * np.array([-1,0,0])) 
        colors = [RED, YELLOW, BLUE, DARK_GRAY]

        vf = ArrowVectorField(func = func, colors=[BLUE])

        plane = ComplexPlane(background_line_style={               
                "stroke_opacity": 0
                           })
        plane.add_coordinates()

        t = ValueTracker(0)

        dot = always_redraw(
            lambda : Dot(point = [np.cos(t.get_value()), np.sin(t.get_value()), 0], color = RED)
        )

        
        self.add(plane)
        self.camera.frame.save_state()
        self.wait()
        self.play(DrawBorderThenFill(vf), run_time = 1.5)
        self.wait()
        self.play(DrawBorderThenFill(dot))
        self.wait()
        self.play(self.camera.frame.animate.scale(0.5), run_time = 2)
        self.wait()
        self.play(t.animate.set_value(2*PI), run_time = 3)
        self.wait()
        
```

```python
%%manim -v WARNING --disable_caching -qh CircleScene

class CircleScene(MovingCameraScene):
    def construct(self):
        plane = ComplexPlane(background_line_style={               
                "stroke_opacity": 0
                           })
        plane.add_coordinates()

        circle = Circle(radius=1, color = BLUE_C)

        t1 = MathTex(r'e^{ix}')
        t1.move_to(plane.c2p(-2,1.5))
        t1.set_color(BLUE_C)

        t2 = MathTex(r'\frac{d}{dx}e^{ix} = ie^ix')
        t2.move_to(plane.c2p(2,1.5))
        t2.set_color(GREEN_C)
        t2.scale(0.7)

        t3 = MathTex(r'|\frac{d}{dx}e^{ix}| = |ie^ix|')
        t3.move_to(plane.c2p(2,1.5))
        t3.set_color(GREEN_C)
        t3.scale(0.7)

        t4 = MathTex(r'= 1')
        t4.move_to(plane.c2p(2.3,1))
        t4.set_color(GREEN_C)
        t4.scale(0.7)

        t = ValueTracker(0)
        dot = always_redraw(
            lambda : Dot(point = [np.cos(t.get_value()), np.sin(t.get_value()), 0], color = RED)
        )

        t5 = MathTex(r'e^{i\pi} = -1')
        t5.move_to(plane.c2p(-2, 1))
        t5.set_color(RED)
        t5.scale(1)

        t6 = MathTex(r'e^{i\pi} + 1 = 0')
        t6.move_to(plane.c2p(-2, 1))
        t6.set_color(RED)
        t6.scale(1)

        t7 = MathTex(r'e^{i\theta} = cos{\theta} + isin{\theta}')
        t7.move_to(plane.c2p(-2, 1))
        t7.set_color(RED)
        t7.scale(0.7)


        # self.add(plane, t1, t5, circle)

        self.add(plane, t1)
        self.camera.frame.save_state()
        self.play(self.camera.frame.animate.scale(0.5), run_time = 2)
        self.play(DrawBorderThenFill(circle))
        self.wait()
        self.play(Write(t2))
        self.wait()
        self.play(Transform(t2, t3))
        self.wait()
        self.play(Write(t4))
        self.wait()
        self.play(DrawBorderThenFill(dot))
        self.wait()
        self.play(t.animate.set_value(PI),Write(t5), run_time = 2)
        self.wait()
        self.play(Transform(t5, t6))
        self.wait()
        self.play(Transform(t5, t7))
        self.wait()
        self.play(t.animate.set_value(2*PI), run_time = 2)
        self.wait()
        self.play(t.animate.set_value(4*PI), run_time = 4)
        self.wait()
```
